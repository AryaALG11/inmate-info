```python
# Drop columns with more than 10% data missing
df.drop(columns=['Next Parole Review Date'], inplace=True)
df
```

```python
#Filling null records
dfd = df.fillna('unknown')
dfd
```

```python
import pandas as pd

# Assuming your DataFrame is named df

# Convert date-related columns to datetime
date_columns = ['Projected Release', 'Maximum Sentence Date', 'Parole Eligibility Date', 
                'Sentence Date', 'Offense Date', 'Last Parole Decision']
df[date_columns] = df[date_columns].apply(pd.to_datetime, errors='coerce')

# Convert Sentence (Years) column to numeric
df['Sentence (Years)'] = pd.to_numeric(df['Sentence (Years)'], errors='coerce')

# Convert categorical columns to categorical data type
categorical_columns = ['Gender', 'Race', 'Current Facility', 'County', 'TDCJ Offense', 'Parole Review Status']
df[categorical_columns] = df[categorical_columns].astype('category')

# Display the data types after conversion
print(df.dtypes)
```

```python
#Removing duplicates
dfd.drop_duplicates(subset='SID Number', inplace=True)
dfd
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ethnicity_counts = dfd['Race'].value_counts()

# Mapping abbreviated ethnicity names to full names
ethnicity_mapping = {
    'H': 'Hispanic',
    'B': 'Black',
    'W': 'White',
    'A': 'Asian',
    'O': 'Other',
    'I': 'Indigenous',
    'U': 'Unknown'
}

# Map the index of the 'ethnicity_counts' Series to their full names
ethnicity_counts.index = ethnicity_counts.index.map(ethnicity_mapping)

# Define colors for each ethnicity
ethnicity_colors = {
    'Hispanic': 'skyblue',
    'Black': 'salmon',
    'White': 'lightgreen',
    'Asian': 'yellow',
    'Other': 'purple',
    'Indigenous': 'orange',
    'Unknown': 'gray'
}

# Plotting
plt.figure(figsize=(10, 6))
ethnicity_counts.plot(kind='bar', color=[ethnicity_colors[ethnicity] for ethnicity in ethnicity_counts.index])

plt.title('Number of Crimes Committed by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)  # Rotate x-labels for better readability

plt.tight_layout()
plt.show()
```

```python
# Iterate through each row in the DataFrame
for index, row in dfd.iterrows():
    # Get the offense code from the current row
    offense_code = row['Offense Code']
    
    # Cut the last three decimal points
    new_offense_code = int(offense_code / 10000)
    
    # Update the 'Offense Code' column with the new value
    dfd.at[index, 'Offense Code'] = new_offense_code

# Display the updated DataFrame
dfd
```

```python
inmate_df = df

# Create groups based on the range of Offense Code
inmate_df['Code Range'] = inmate_df['Offense Code'] // 1000000

# Map common words from TDCJ Offense to each group
code_ranges = inmate_df.groupby('Code Range')['TDCJ Offense'].apply(lambda x: x.str.split().str[0].mode().iloc[0])

# Rename the groups starting from 1, 2, 3 in serial order
code_ranges = code_ranges.reset_index()
code_ranges['Code Range'] = range(1, len(code_ranges) + 1)

# Display the results
for row in code_ranges.itertuples(index=False):
    print(f"Group {row[0]}: {row[1]} Activity")
    
```

```python
# 1. Create a dictionary to map group number to its corresponding offense
grouped_offenses_dict = dict(zip(code_ranges['Code Range'], code_ranges['TDCJ Offense']))

# 2. Append a new row to the code_ranges DataFrame
new_row = {'Code Range': len(code_ranges) + 1, 'TDCJ Offense': 'Other Offenses'}
code_ranges = code_ranges.append(new_row, ignore_index=True)

# 3. Merge the code_ranges DataFrame with the inmate_df DataFrame
inmate_df = pd.merge(inmate_df, code_ranges, on='Code Range', how='left')

# Display the first few rows of the merged DataFrame
inmate_df
```
